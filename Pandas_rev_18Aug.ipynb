```python
data4={
    'emp_id':[101,102,103,104],
    'salary':[234,452,333,523],
    'pf':[23,41,47,34]
}
```

```python
data5={
    'emp_id':[101,102,104,105],
    'mob_no':[1234,452,2345,4232],
    'house_no':[123,343,465,462]
}
```

```python
d5=pd.DataFrame(data4)
d6=pd.DataFrame(data5)
```

```python
pd.merge(d5,d6) #inner join
```

```python
# right join
pd.merge(d5,d6,how='right',on='emp_id')
```

```python
# left jon
pd.merge(d5,d6,how='left',on='emp_id')
```

```python
pd.merge(d5,d6,left_on='emp_id',right_on='emp_id')
```
